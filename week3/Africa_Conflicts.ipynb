```python
#Time series analysis
import pandas as pd
import matplotlib.pyplot as plt

# Grouping by year
events_by_year = df.groupby('YEAR').size()
fatalities_by_year = df.groupby('YEAR')['FATALITIES'].sum()

#plotting the events and fatalities over time
plt.figure(figsize=(12,6))
plt.plot(events_by_year.index, events_by_year.values, label='Number of Events', marker='o')
plt.plot(fatalities_by_year.index, fatalities_by_year.values, label='Total Fatalities', marker='o')
plt.title('Confict Events and Fatalities Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid()
plt.show()
```

```python
print(df[['LATITUDE','LONGITUDE']].dtypes)
```

```python
df['LONGITUDE'] = df['LONGITUDE'].str.replace(',','.').astype(float)
```

```python
#geospatial analysis
import folium 

#creating a map centered on Africa
m = folium.Map(location=[8,20], zoom_start=4)

# adding conflict events as markers
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        radius=row['FATALITIES']/10, #scaling fatalities for visibility
        color='red',
        fill=True,
        fill_color='red',
        popup=f"Event: {row['EVENT_TYPE']}, Fatalities: {row['FATALITIES']}"
    ).add_to(m)

# saving and displaying the map
m.save('conflict_map.html')
```

```python
#analyzing different actors in conflict

#top 10 actors involved in conflicts
top_actors = df['ACTOR1'].value_counts(10)
print(top_actors)
```

```python
#fatalities by actor
fatalities_by_actor = df.groupby('ACTOR1')['FATALITIES'].sum().sort_values(ascending=False).head(10)
print(fatalities_by_actor)
```

```python
#plotting the fatalities bt actor
fatalities_by_actor.plot(kind='bar', figsize=(10,6), title='Top 10 Actors by Fatalities')

plt.xlabel('Actor')
plt.ylabel('Total Fatalities')
plt.show()
```

```python
#fatality analysis

# deadliest event types
deadliest_event_types = df.groupby('EVENT_TYPE')['FATALITIES'].sum().sort_values(ascending=False)
print(deadliest_event_types)
```

```python
deadliest_regions = df.groupby('REGION')['FATALITIES'].sum().sort_values(ascending=False)
print(deadliest_regions)
```

```python
# boxplot of fatalities by event type
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.boxplot(x='EVENT_TYPE', y='FATALITIES', data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Fatalities by Event Type')
plt.show()
```

```python
# distribution of event types
event_type_distribution = df['EVENT_TYPE'].value_counts()
print(event_type_distribution)
```

```python
# fatalities by sub-event type
fatalities_by_sub_event= df.groupby('SUB_EVENT_TYPE')['FATALITIES'].sum().sort_values(ascending=False)

print(fatalities_by_sub_event)
```

```python
# plotting the event type distribution
event_type_distribution.plot(kind='bar', figsize=(12, 6), title='Distribution of Event Types')
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.show()
```

```python
# correlation between variables
import seaborn as sns

#creating a pivot table of fatalities bt event type and year
pivot_table = df.pivot_table(values='FATALITIES', index='YEAR', columns='EVENT_TYPE', aggfunc='sum')

#plotting a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlOrRd', annot=True, fmt='.0f')
plt.title('Fatalities by Event Type and Year')
plt.show()
```
